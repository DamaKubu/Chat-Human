### Collapsible induction task generation code

```python
import random

def generate_example(num_digits=5):
    number = str(random.randint(10**(num_digits-1), (10**num_digits)-1))
    base_choices = ["first", "second", "third", "fourth", "fifth", "sixth", "seventh", "eighth", "nineth", "tenth"]
    choices = base_choices[:num_digits]
    position = random.choice(choices)

    digit_index = {"first": 0, "second": 1, "third": 2, "fourth": 3, "fifth": 4, "sixth": 5, "seventh": 6, "eighth": 7, "nineth": 8, "tenth": 9}
    digit = int(number[digit_index[position]])
    sentence = f"The {position} digit of {number} is {digit}. "

    return sentence

prompt = ""
for _ in range(5):
    example = generate_example(6)
    prompt += example

prompt = prompt.strip()
prompt
```

### Mamba Interp

```python
prompt = "The third digit of 566406 is 6. The second digit of 521808 is 2. The sixth digit of 122657 is 7. The first digit of 156443 is"
tokenized_prompt = tokenizer.tokenize(prompt)
print(len(tokenized_prompt), tokenized_prompt)
print(tokenized_prompt[31])
```

This prompt is 38 tokens long. Let's intervene on " first" at position 31 to narrow down the inner components responsible for processing this information. In a Mamba model, there are two ways to do this:

1. We can patch the activations from a prompt with a separate token e.g. " second" (Does not involve recurrent dimension)
2. Intervene on the SSM block hidden state (Involves recurrent dimension)

See the nnsight docs for how to do (1) (there's also a section below)

For (2), here's how you can access those recurrent hidden states.

#### Getting Values

https://nnsight.net/notebooks/features/getting/

```python
from transformers import AutoTokenizer
from nnsight.models.Mamba import MambaInterp

tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-neox-20b", padding_side="left")
tokenizer.pad_token_id = tokenizer.eos_token_id
mamba_model = MambaInterp("state-spaces/mamba-130m", device="cuda", tokenizer=tokenizer)

prompt = "The third digit of 566406 is 6. The second digit of 521808 is 2. The sixth digit of 122657 is 7. The first digit of 156443 is"
layer_num = 10

with mamba_model.invoke(prompt) as invoker:
    # We hook the hidden state
    hx = mamba_model.backbone.layers[layer_num].mixer.ssm.hx
    # Then access its value. You can also access module inputs with .input
    h1 = hx.output.save()
    # Since we're accessing the output of hx from the 0th position, this gives the hidden state at pos 1
    # Like a linked list, we can call hx.next() to move to the next position.
    h2 = hx.next().output.save()
    h3 = hx.next().output.save()
    h4 = hx.next().output.save()

print(hx, type(hx))
print(hx.output) #this doesn't have anything it's just the proxy
print(h1.value.shape) # this gets the shape
print(h1.value) # this gets the value
#print(hx.output.save().shape) # you can't save outside of the invoker (it will not give an error but it's still a proxy)
```

#### Setting Values

https://nnsight.net/notebooks/features/setting/

```python
import torch
with mamba_model.invoke(prompt) as invoker:
    hx = mamba_model.backbone.layers[layer_num].mixer.ssm.hx
    hidden_states_pre = hx.output.clone().save()

    noise = (0.001**0.5)*torch.randn(hidden_states_pre.shape)
    hx.output[0][:] = hidden_states_pre + noise

    hidden_states_post = hx.output[0].save()
```

```python
# We can see the change in the results
print(hidden_states_pre.value)
print(hidden_states_post.value)
```

#### Intervening on the SSM

```python
with mamba_model.invoke(prompt) as invoker:
    hx = mamba_model.backbone.layers[layer_num].mixer.ssm.hx
    h1 = hx.output.clone().save()

    # we want to access and intervene at position 31
    for i in range(29):
        hx.next()

    h30 = hx.output.clone().save()
    hx.next()
    h31_pre = hx.output.clone().save()
    hx.next()

    # setting h31 to be equal to h30
    hx.output[0][:] = h30

    h31_post = hx.output.save()
```

```python
# We can see the change in the results
print(h30.value)
print(h31_pre.value)
print(h31_post.value)
print((h30.value == h31_post.value).all())
```

#### (1) Reminder on how activation patching works in nnsight

Source: https://nnsight.net/notebooks/tutorials/ioi_patching/

```python
from nnsight import LanguageModel, util
from nnsight.tracing.Proxy import Proxy

# We load in a TinyStories llama model
llama_model = LanguageModel("delphi-suite/delphi-llama2-100k", device_map="cuda:0")
# Set the clean and corrupt prompts
clean_prompt = "Once upon a time, John and Mary went shopping. After John and Mary went to the store, Mary gave a bottle of milk to"
corrupted_prompt = (
    "Once upon a time, John and Mary went shopping. After John and Mary went to the store, John gave a bottle of milk to"
)

# Get the vocabulary indices for the correct and incorrect prediction
correct_index = llama_model.tokenizer("John")["input_ids"][1]
incorrect_index = llama_model.tokenizer("Mary")["input_ids"][1]

print(f"' John': {correct_index}")
print(f"' Mary': {incorrect_index}")
```

```python
# Enter nnsight tracing context
with llama_model.forward() as runner:

    # Clean run
    with runner.invoke(clean_prompt) as invoker:
        clean_tokens = invoker.input["input_ids"][0]

        # Get hidden states of all layers in the network.
        # We index the output at 0 because it's a tuple where the first index is the hidden state.
        # No need to call .save() as we don't need the values after the run, just within the experiment run.
        clean_hs = [llama_model.model.layers[layer_idx].output[0] for layer_idx in range(len(llama_model.model.layers))]

        # Get logits from the lm_head.
        clean_logits = llama_model.lm_head.output

        # Calculate the difference between the correct answer and incorrect answer for the clean run and save it.
        clean_logit_diff = (clean_logits[0, -1, correct_index] - clean_logits[0, -1, incorrect_index]).save()

    # Corrupted run
    with runner.invoke(corrupted_prompt) as invoker:
        corrupted_logits = llama_model.lm_head.output

        # Calculate the difference between the correct answer and incorrect answer for the corrupted run and save it.
        corrupted_logit_diff = (corrupted_logits[0, -1, correct_index] - corrupted_logits[0, -1, incorrect_index]).save()

    ioi_patching_results = []

    # Iterate through all the layers
    for layer_idx in range(len(llama_model.model.layers)):
        _ioi_patching_results = []

        # Iterate through all tokens
        for token_idx in range(len(clean_tokens)):

            # Patching corrupted run at given layer and token
            with runner.invoke(corrupted_prompt) as invoker:

                # Apply the patch from the clean hidden states to the corrupted hidden states.
                llama_model.model.layers[layer_idx].output[0].t[token_idx] = clean_hs[layer_idx].t[token_idx]

                patched_logits = llama_model.lm_head.output

                patched_logit_diff = (patched_logits[0, -1, correct_index] - patched_logits[0, -1, incorrect_index])

                # Calculate the improvement in the correct token after patching.
                patched_result = (patched_logit_diff - corrupted_logit_diff) / (clean_logit_diff - corrupted_logit_diff)

                _ioi_patching_results.append(patched_result.save())

        ioi_patching_results.append(_ioi_patching_results)
```

```python
import plotly.express as px
print(f"Clean logit difference: {clean_logit_diff.value:.3f}")
print(f"Corrupted logit difference: {corrupted_logit_diff.value:.3f}")

ioi_patching_results = util.apply(ioi_patching_results, lambda x: x.value.item(), Proxy)

#clean_tokens = [llama_model.tokenizer.decode(token) for token in clean_tokens]
#token_labels = [f"{token}_{index}" for index, token in enumerate(clean_tokens)]

fig = px.imshow(
    ioi_patching_results,
    color_continuous_midpoint=0.0,
    color_continuous_scale="RdBu",
    labels={"x": "Position", "y": "Layer"},
    #x=token_labels, # doesn't work for some reason
    title="Normalized Logit Difference After Patching Residual Stream on the IOI Task",
)

fig.show()
```

#### (1*) Activation patching in Mamba

```python
# We load in a TinyStories llama model
mamba_model = MambaInterp("MrGonao/delphi-mamba-100k", device="cuda")
# Set the clean and corrupt prompts
clean_prompt = "Once upon a time, John and Mary went shopping. After John and Mary went to the store, Mary gave a bottle of milk to"
corrupted_prompt = (
    "Once upon a time, John and Mary went shopping. After John and Mary went to the store, John gave a bottle of milk to"
)

# Get the vocabulary indices for the correct and incorrect prediction
correct_index = mamba_model.tokenizer("John")["input_ids"][1]
incorrect_index = mamba_model.tokenizer("Mary")["input_ids"][1]

print(f"' John': {correct_index}")
print(f"' Mary': {incorrect_index}")
```

```python
# Indicate we want to run the model.
with mamba_model.forward(validate=False) as runner:
    # Indicate we want to enter some input into the model.
    # Run clean prompt though in order to grab the clean hidden states.
    with runner.invoke(clean_prompt, scan=False) as invoker:
        clean_tokens = invoker.input["input_ids"][0]

        # Grab hidden states as output of each layer module.
        clean_hs = [
            mamba_model.backbone.layers[layer_idx].output
            for layer_idx in range(len(mamba_model.backbone.layers))
        ]

        # Grab clean logits
        clean_logits = mamba_model.lm_head.output

        # Compute logit diff between correct and incorrect indices.
        clean_logit_diff = (
            clean_logits[0, -1, correct_index] - clean_logits[0, -1, incorrect_index]
        ).save()

    # Run a single corrupted prompt to get logit diff of corrupted.
    with runner.invoke(corrupted_prompt, scan=False) as invoker:
        corrupted_logits = mamba_model.lm_head.output

        corrupted_logit_diff = (
            corrupted_logits[0, -1, correct_index]
            - corrupted_logits[0, -1, incorrect_index]
        ).save()

    ioi_patching_results = []

    # Iterate through each layer.
    for layer_idx in range(len(mamba_model.backbone.layers)):
        _ioi_patching_results = []

        # Iterate through each token.
        for token_idx in range(len(clean_tokens)):
            with runner.invoke(corrupted_prompt, scan=False) as invoker:
                # Patch in hidden states from clean run into corrupted at given layer and token.
                mamba_model.backbone.layers[layer_idx].output[0][:, token_idx] = clean_hs[
                    layer_idx
                ][0][:, token_idx]
                # Output of each layer is a tuple of (hidden_states, residual). Need to patch both.
                mamba_model.backbone.layers[layer_idx].output[1][:, token_idx] = clean_hs[
                    layer_idx
                ][1][:, token_idx]

                patched_logits = mamba_model.lm_head.output

                patched_logit_diff = (
                    patched_logits[0, -1, correct_index]
                    - patched_logits[0, -1, incorrect_index]
                )

                # Compute improvement in logit diff.
                patched_result = (patched_logit_diff - corrupted_logit_diff) / (
                    clean_logit_diff - corrupted_logit_diff
                )

                _ioi_patching_results.append(patched_result.save())

        ioi_patching_results.append(_ioi_patching_results)


print(f"Clean logit difference: {clean_logit_diff.value:.3f}")
print(f"Corrupted logit difference: {corrupted_logit_diff.value:.3f}")

# Convert proxies to their value.
ioi_patching_results = util.apply(ioi_patching_results, lambda x: x.value.item(), Proxy)
```

```python
import plotly.express as px

clean_tokens = [mamba_model.tokenizer.decode(token) for token in clean_tokens]
token_labels = [f"{token}_{index}" for index, token in enumerate(clean_tokens)]

fig = px.imshow(
    ioi_patching_results,
    color_continuous_midpoint=0.0,
    color_continuous_scale="RdBu",
    labels={"x": "Position", "y": "Layer"},
    x=token_labels,
    title="Normalized Logit Difference After Patching Residual Stream on the IOI Task in Mamba",
)

fig.show()
```
